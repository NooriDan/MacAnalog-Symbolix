# PySpice Directly with Skywater

```python
from PySpice.Spice.Library import SpiceLibrary
from PySpice.Spice.Netlist import Circuit
from PySpice.Probe.Plot import plot
from PySpice.Unit import *
import numpy as np

SKYWATER_PATH       = "../../skywater-pdk"
DEVICE_INFO_FOLDER  = SKYWATER_PATH + "/libraries/sky130_fd_pr/latest/cells"
LIB_PATH            = SKYWATER_PATH + "/libraries/sky130_fd_pr/latest/models/sky130.lib.spice"
# LIB_PATH            = SKYWATER_PATH + "/libraries/sky130_fd_pr/latest/models"
```

```python
# spice_library = SpiceLibrary('/path/to/skywater/pdk/spice_models')
spice_library = SpiceLibrary(LIB_PATH,
                             recurse=True,
                             section='tt')

spice_library = SpiceLibrary(DEVICE_INFO_FOLDER,
                             recurse=True)
```

```python
print(f"loaded: {len(list(spice_library.models))}\n{list(spice_library.models)}\n")
```

```python
circuit = Circuit('Simple Amplifier')

# Add components
circuit.V(1, 'Vdd', circuit.gnd, 5@u_V)  # Voltage source
circuit.R(1, 'out', circuit.gnd, 1@u_kΩ) # Resistor
circuit.M(1, 'out', 'gate', circuit.gnd, circuit.gnd, model='nmos_model')

# Include SPICE models from the Skywater library
circuit.include(spice_library['sky130_fd_pr__diode_pw2nd_11v0'])
```

```python
from PySpice.Spice.Parser import SpiceParser

parser = SpiceParser('/path/to/netlist.cir')
circuit = parser.build_circuit()
```

```python
from PySpice.Spice.Simulation import DcAnalysis

# Example: Sweeping a resistor value
resistor_values = np.linspace(1, 10, 10) * u_kΩ  # 1kΩ to 10kΩ

for r_value in resistor_values:
    circuit.R(1, 'out', circuit.gnd, r_value)  # Update resistor value

    # Run a DC analysis
    simulator = circuit.simulator(temperature=25, nominal_temperature=25)
    analysis = simulator.dc(Vdd=slice(0, 5, 0.1))  # Sweep Vdd from 0 to 5V

    # Extract and analyze results
    for node in analysis.nodes.values():
        print(f"{node} = {float(node)}")
```

```python
from PySpice.Spice.Simulation import TransientAnalysis

simulator = circuit.simulator(temperature=25, nominal_temperature=25)
analysis = simulator.transient(step_time=1@u_us, end_time=1@u_ms)

# Plot results
plot(analysis['out'], title="Transient Analysis")
```

```python
from PySpice.Spice.Simulation import AcAnalysis

simulator = circuit.simulator(temperature=25, nominal_temperature=25)
analysis = simulator.ac(start_frequency=10@u_Hz, stop_frequency=1@u_MHz, number_of_points=100)

# Plot results
plot(analysis['out'], title="AC Analysis")
```

```python
# Vary W (width) and L (length) for an NMOS
width_values = [1, 2, 3]  # micrometers
length_values = [0.18, 0.35]  # micrometers

for W in width_values:
    for L in length_values:
        circuit.M(1, 'out', 'gate', circuit.gnd, circuit.gnd,
                  model='sky130_fd_pr__nfet_01v8',
                  w=W * u_um, l=L * u_um)

        simulator = circuit.simulator(temperature=25, nominal_temperature=25)
        analysis = simulator.dc(Vdd=slice(0, 5, 0.1))
```

```python
import matplotlib.pyplot as plt

# Example: Plotting results
plt.plot(analysis.sweep, analysis['out'])
plt.title('DC Analysis')
plt.xlabel('Vdd (V)')
plt.ylabel('Output Voltage (V)')
plt.grid()
plt.show()
```
